```python
## Scrip to record a click on any ad and stram the json click data to kafka streaming with the topis as publisher id 


#Creating fake data 
import json
import time
import random
import datetime
from faker import Faker


fake = Faker()

#Creating dummy advertiser id data
advertiser_id=[]
for i in range (21,41):
    advertiser_id.append('Adv_0'+str(i))
    
ad_type = ['Image', 'Video', 'Pop Up']

#Creating dummy publisher id data
publisher_id=[]
for i in range (1,11):
    if i < 10:
        publisher_id.append('Pb_00'+str(i))
    else:
        publisher_id.append('Pb_0'+str(i))
        
#Creating dummy ad id data
ad_id=[]
for i in range (1,101):
    if i < 10:
        ad_id.append('Ad_00'+str(i))
    elif i < 100:
        ad_id.append('Ad_0'+str(i))
    else:
        ad_id.append('Ad_'+str(i))

#Creating dummy exhange id data
exchange_id = ['NULL']
for i in range(5):
    exchange_id.append(random.choice(publisher_id)+ '_'+
                       random.choice(publisher_id))

#creating dummy impression data
impression = [  ] 
for i in range (1,101):
    impression.append(i)
    

#Creating dummy CTR data
ctr=['YES','NO']
        


```

```python

#Intializing kafka Message Producer
from kafka import KafkaProducer
import json

class MessageProducer:
    broker = ""
    topic = ""
    producer = None

    def __init__(self, broker, topic):
        self.broker = broker
        self.topic = topic
        self.producer = KafkaProducer(bootstrap_servers=self.broker,
        value_serializer=lambda v: json.dumps(v).encode('utf-8'),
        acks='all',
        retries = 3)


    def send_msg(self, msg):
        print("sending message...")
        try:
            future = self.producer.send(self.topic,msg)
            self.producer.flush()
            future.get(timeout=60)
            print("message sent successfully...")
            return {'status_code':200, 'error':None}
        except Exception as ex:
            return ex


broker = 'localhost:9092'
topic = 'Topic1'

# message_producer = MessageProducer(broker,topic)

# data = generate_json_object()
# resp = message_producer.send_msg(data)
# print(resp)
```

```python
#Creating a json object with dummy data and publishing it to kafka with topic as publisher id

def generate_json_object():    
    


    dic = {
         "timestamp": int(time.time()*1000),
         "publisher_id":  random.choice(publisher_id),
         "ad_id": random.choice(ad_id),
         "adv_id": random.choice(advertiser_id),
         "exchange_id": random.choice(exchange_id),
         "ad_type": random.choice(ad_type),
         "impersion": random.choice(impression),
         "first_ctr": random.choice(ctr)

    }


    broker = 'localhost:9092'
    topic = dic["publisher_id"]

    message_producer = MessageProducer(broker,topic)


    resp = message_producer.send_msg(dic)
    
    return 0

```

```python
## Intializing a simple flask application with dash lib with a simple button.


from dash import Dash, html, Input, Output, ctx

app = Dash(__name__)



app.layout = html.Div([
    html.Button('Button', id='btn-nclicks-1', n_clicks=0),
    html.Div(id='container-button-timestamp')
])



# Callback to trigger message brodcast when button is clicked
@app.callback(
    Output('container-button-timestamp', 'children'),
    Input('btn-nclicks-1', 'n_clicks')
)
def displayClick(btn1):    #function to call json object and push it to kafka
    msg = "None of the buttons have been clicked yet"
    if "btn-nclicks-1" == ctx.triggered_id:
        msg = "Message genrated"
        generate_json_object()
    
        
    return html.Div(msg)



if __name__ == '__main__':
    app.run_server(debug=True, use_reloader=False)
```
